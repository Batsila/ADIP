```python
from scipy import ndimage
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
gray_image = misc.face(gray=True)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.imshow(gray_image, cmap=plt.cm.gray)
ax1.axis('off')
ax1.set_title('Input gray Image', fontsize=20)

ax2.hist(gray_image.ravel(), 256, [0,256])
ax2.set_title('Gistogram', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
import PIL
from PIL import Image

def hexencode(rgb):
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    return '#%02x%02x%02x' % (r,g,b)

def get_main_color(colors, min_hits):
    average_colors = []
    for c in colors:
        if(c[0] > min_hits):
            average_colors.append(c)

    return average_colors
```

```python
color_image = Image.open('./test.jfif')  

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.imshow(color_image)
ax1.axis('off')
ax1.set_title('Input color Image', fontsize=20)

ax2.set_title('Gistogram', fontsize=20)

w, h = color_image.size  
colors = color_image.getcolors(w * h)
main_colors = get_main_color(colors, 15)

for idx, c in enumerate(main_colors):
    ax2.bar(idx, c[0], color=hexencode(c[1]),edgecolor=hexencode(c[1]))

fig.tight_layout()
plt.show()
```
